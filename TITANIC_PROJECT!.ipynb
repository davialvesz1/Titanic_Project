```python
### Objetivo deste projeto:Encontrar o nome das 10 sobreviventes mais jovens da primeira classe do Titanic.
```

```python
#Importar bibliotecas
import pandas as pd

```

```python
#carregar dadaos
df = pd.read_csv('titanic.csv')
```

```python
#filtrar determinadas colunas
filter1 = df[['Name','Sex','Age','Pclass','Survived']]
```

```python
#filtrar apenas pessoas do sexo feminino
filter1 = filter1.loc[filter1["Sex"]=="female"]
#filtrar pessoas do sexo feminino que estavam na primeira classe
filter1 = filter1.loc[filter1['Pclass']==1]
#filtrar apenas pessoas do sexo feminino da primeira classe que sobreviveram
filter1 = filter1.loc[filter1['Survived']==1]
#organizar por idade da mais nova para a mais velha
filter1 = filter1.sort_values(['Age'],ascending=True)
```
